```python
import numpy as np

N = 1000

x = np.random.randn(N)
y = np.random.randn(N)
```

```python
s = np.sqrt(1/2) * (x + 1j*y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(s.real, s.imag)
```

```python
# Melhorando qualidade da simulação

plt.figure(figsize=(6,6))

plt.scatter(s.real, s.imag, s=10)

plt.xlabel("In-Phase (Real)")
plt.ylabel("Quadrature (Imaginary)")

plt.title("Scatter Plot do Sinal Complexo")

plt.grid(True)

plt.axis("equal")

plt.show()
```
