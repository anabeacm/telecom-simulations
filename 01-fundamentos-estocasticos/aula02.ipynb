```python
# Simulação de sinal
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 500)

sinal = np.sin(t)

plt.plot(t, sinal)
plt.grid()
plt.show()
```

```python
ruido = np.random.randn(500)
```

```python
plt.plot(t, ruido)
plt.grid()
plt.show()
```

```python
recebido = sinal + 0.2*ruido
```

```python
plt.plot(t, recebido)
plt.grid()
plt.show()
```
